```python
#simple recommender system using average weights using the vote average and popularity
import numpy as np
import pandas as pd
```

```python
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
```

```python
#Combining the two data frames based on the id's column as they have the same value
credits_rename=credits.rename(index=str,columns={"movie_id":"id"})
credits_rename.head()
movies_merge=movies.merge(credits_rename,on='id')
movies_merge.head()
# movies_merge.shape
```

```python
movies_merge.shape
```

```python
movies_clean=movies_merge.drop(columns=['homepage','title_x','title_y','status','production_countries'])
movies_clean.head()
```

```python
movies_clean.shape
```

```python
#info about the data set
movies_clean.info()
```

No null value observed
W=(R*v+C*m)/(v+m)
here W=Weighted rating
R=average for the movies between 0 to 10
v=number of votes
m=minimun votes required to be listed in the top 250
C=the mean vote across the whole report

```python
#claculation of components
v=movies_clean['vote_count']
R=movies_clean['vote_average']
C=movies_clean['vote_average'].mean()
m=movies_clean['vote_count'].quantile(0.70)
```

```python
movies_clean['weighted_avg']=((R*v)+(C*m))/(v+m)
```

```python
movies_clean.head()
```

```python
movie_sorted_ranking=movies_clean.sort_values('weighted_avg',ascending=False)
movie_sorted_ranking[['original_title','vote_count','vote_average','weighted_avg','popularity']].head(20)
```

```python
#This is based on the content based filtering
import matplotlib.pyplot as plt
import seaborn as sns

weight_avg=movie_sorted_ranking.sort_values('weighted_avg',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_avg['weighted_avg'].head(10),y=weight_avg['original_title'].head(10))
plt.title('Best Movies by average votes',weight='bold')
plt.xlabel('Weighted average score',weight='bold')
plt.ylabel('Movie title',weight='bold')
plt.show()
```

```python
popularity=movie_sorted_ranking.sort_values('popularity',ascending=False)

ax=sns.barplot(x=popularity['popularity'].head(10),y=popularity['original_title'].head(10))

plt.title('Most popular by votes',weight='bold')
plt.xlabel('Score of popularity',weight='bold')
plt.ylabel('Movie title',weight='bold')
plt.show()
```

```python
#Recommendation based on sclar wighted average and popularity socre(priority is 50-50)
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
movie_scaled=scale.fit_transform(movies_clean[['weighted_avg','popularity']])
movie_normalized=pd.DataFrame(movie_scaled,columns=['weighted_avg','popularity'])
movie_normalized.head()
#normalizing  the values of weighted avg and popularit between 0 to 1
```

```python
movies_clean[['normaized_weight','normalized_popularity']]=movie_normalized
movies_clean.head()
```

```python
movies_clean['score']=movies_clean['normaized_weight']*0.5+movies_clean['normalized_popularity']*0.5
#giving the priority score
movies_sorced=movies_clean.sort_values(['score'],ascending=False)
movies_sorced[['original_title','normaized_weight','normalized_popularity','score']].head()
```

```python
scored=movies_clean.sort_values('score',ascending=False)

ax=sns.barplot(x=scored['score'].head(10),y=scored['original_title'].head(10))

plt.title('Best rated & Most popular blend',weight='bold')
plt.xlabel('Score',weight='bold')
plt.ylabel('Movie Title',weight='bold')
plt.show()
```
